## Build Dataset

```python
history_df = review_df.rename(columns={"timestamp": "time"})
history_df = history_df[["user_id", "parent_asin", "time"]]
history_df["time"] = history_df["time"].astype(int)
history_df.drop_duplicates(inplace=True)
history_df = history_df.sort_values(by=["time", "user_id"], kind="mergesort").reset_index(drop=True)
history_df.head(n=5)
```

```python
uid_set = sorted(history_df["user_id"].unique())
uid2id = {uid: id for id, uid in enumerate(uid_set)}
iid_set = sorted(history_df["parent_asin"].unique())
iid2id = {iid: id for id, iid in enumerate(iid_set)}

history_df["user_id"] = history_df["user_id"].apply(lambda x: uid2id[x])
history_df["parent_asin"] = history_df["parent_asin"].apply(lambda x: iid2id[x])
history_df.head(n=5)
```

```python
# leave one out evaluation 留一法评估，用户序列最后一个和倒数第二个 item 取出，作为 eval 和 test 的正样本，其余作为 train

# 用户点击合集，用于后续的负采样
user_click_item_set = dict()
for user_id, item_seq in history_df.groupby("user_id"):
    user_click_item_set[user_id] = set(item_seq["parent_asin"].values.tolist())

def gen_3_set(df):
    eval_test = []
    item_num = df["parent_asin"].value_counts().size
    df = df.groupby("user_id").filter(lambda x: len(x) >= 3)
    for i in range(2):
        temp_df = df.groupby("user_id").tail(1).copy()
        df = df.drop(temp_df.index)
        neg_items = np.random.randint(0, item_num - 1, size=(len(temp_df), NEG_ITEM))
        for i, uid in enumerate(temp_df["user_id"].values):
            user_clicked = user_click_item_set[uid]
            for j in range(len(neg_items[i])):
                while neg_items[i][j] in user_clicked:
                    neg_items[i][j] = np.random.randint(0, item_num - 1)
        temp_df["neg_items"] = neg_items.tolist()
        eval_test.append(temp_df)
    return eval_test, df
```

```python
# 保底每个用户都有一个交互记录
guarantee_set = history_df.groupby("user_id").head(1)
history_df = history_df.drop(guarantee_set.index)

[eval_data, test_data], train_data = gen_3_set(history_df)
train_data = pd.concat([train_data, guarantee_set]).sort_index()

print(f"train data: {len(train_data)}, eval data: {len(eval_data)}, test data: {len(test_data)}")
```

### Item data

```python
item_df = interact_meta_df[[
    "title", 
    "average_rating",
    "rating_number",
    "features",
    "description",
    "price",
    "categories", 
    "details",
    "parent_asin"
    ]]
item_df["parent_asin"] = item_df["parent_asin"].apply(lambda x: iid2id[x])
item_df["l1_category"] = item_df["categories"].apply(lambda x: x[0] if len(x) > 0 else "")
item_df["l2_category"] = item_df["categories"].apply(lambda x: x[1] if len(x) > 1 else "")
item_df["l3_category"] = item_df["categories"].apply(lambda x: x[2] if len(x) > 2 else "")
item_df.drop(columns=["categories"], inplace=True)
item_df["features"] = item_df["features"].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
item_df["description"] = item_df["description"].apply(lambda x: " ".join(x) if isinstance(x, list) else "")
# convert dict to string
item_df["details"] = item_df["details"].apply(lambda x: " ".join([f"{k}: {v}" for k, v in x.items()]) if isinstance(x, dict) else "")
item_df.head(n=5)
```

```python
item_df.to_csv(os.path.join(ROOT_PATH, "item_df.csv"), index=False)
```
